# **Foreign Exchange Earnings from Tourism in India during 1991-2020**
https://data.gov.in/resource/foreign-exchange-earnings-tourism-india-during-1991-2020
Link for the dataset

```python
df1=pd.read_csv("/content/India-Tourism-Statistics-2021-Table-2.11.1.csv")
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1.info()
```

```python
# Deleting first row
df1=df1.dropna(axis=0)
```

```python
# Cleaning Year column
df1['Year']=df1['Year'].replace('2016#2',"2016")
df1['Year']=df1['Year'].replace('2017#2',"2017")
df1['Year']=df1['Year'].replace('2018#2',"2018")
df1['Year']=df1['Year'].replace('2019#2',"2019")
df1['Year']=df1['Year'].replace('2020#2',"2020")
```

Converting data type of Year column to int as its datatype was object.

```python
df1['Year']=df1['Year'].astype(str)
```

```python
df1['Year']=df1['Year'].astype(int)
```

```python
df1['Year'].dtype
```

Line chart showing annual growwth of Foreign Exchange Earning Year wise.

```python
fig=px.line(df1,x='Year',y='FEE in ` terms - `Crore',text="Year",title="Foreign Exchange year vise",markers=True)
fig.update_traces(textposition="bottom right")
fig.show()
```

from the above line chart we can conclude that their was a good rate of growth of FEE but in the year 2020 the growth rate decreased for which the reason can be Covid-19

```python
import plotly.graph_objects as go
```

```python
fig=go.Figure()
```

```python
fig=px.bar(df1,x='Year',y='FEE in ` terms - % Change over previous year',text="Year",title="Foreign Exchange Change Trend line",color='Year')
fig.show()
```

there was a steap downfall in the trend of foreign exchange earnings after year 2019 as Covid 19 can be the reason and 2003 was the year which had the highest rate of growth in FEE which was never seen again.
